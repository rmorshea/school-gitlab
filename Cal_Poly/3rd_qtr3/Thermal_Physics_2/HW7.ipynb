```python
import numpy as np
import scipy.integrate as integrate
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
rho=8.993e3#kg/m^3
m=1e-25 #kg
Cs=3560 #m/s
h=6.626e-34 #m^2*kg/s
k=1.38e-23 #m^2*kg/s^2
```

###7.58

```python
Td=(h*Cs)/(2*k)*(6*rho/np.pi/m)**(1./3)
print Td
```

###7.60

```python
def func(x):
    return x**4*np.exp(x)/(np.exp(x)-1)**2
def calc_Cv(T):
    return T**3*integrate.quad(func,0.00001,1./T)[0]/Td**3

Td=465.9
ef=1.28e-18
```

```python
T=np.linspace(1,5,1000)
dm_Cv=[]
lv_Cv=[]
ce_Cv=[]
tot=[]
for i in T:
    dm_Cv.append(9*calc_Cv(i))
    lv_Cv.append(12*np.pi**4/5/Td**3*i**3)
    ce_Cv.append(np.pi**2*k/2/ef*i)
    tot.append(np.pi**2*k/2/ef*i+12*np.pi**4/5/Td**3*i**3)
    
plt.figure(figsize=(10,5))
#plt.plot(T,dm_Cv,label='numerical integral')
plt.plot(T,lv_Cv,label='latice vibrations')
plt.plot(T,ce_Cv,label='conduction electrons')
plt.plot(T,tot,label='total')
plt.xlabel('$T/T_D$',fontsize=15)
plt.ylabel('$C_v/Nk$',fontsize=15)
plt.legend(loc='best')
plt.show()
```

```python
def calc_Cv(T):
    return T**2*integrate.quad(lambda x: x**3*np.exp(x)/(np.exp(x)-1)**2,0.1,1./T)[0]
```

```python
T=np.linspace(0.01,1,1000)
Cv=[]
for i in T:
    Cv.append(calc_Cv(i))
    
plt.figure(figsize=(10,5))
plt.plot(T,Cv,label='numerical integral')
plt.xlabel('$T/T_D$',fontsize=15)
plt.ylabel('$C_v/Nk$',fontsize=15)
plt.legend(loc='best')
plt.show()
```
