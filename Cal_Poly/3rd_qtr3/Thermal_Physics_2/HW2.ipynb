```python
e_mu=np.linspace(-60,60,1000)
# x = epsilon - mu
k=8.6e-5
kT= lambda (T): 8.6e-5*T
def n(x,T):
    return (np.exp(-x/kT(T))+np.exp(-2*x/kT(T)))/(1+np.exp(-x/kT(T))+np.exp(-2*x/kT(T)))
def n_bar(T): 
    return [n(x,T) for x in e_mu]
T=[1000,10000,50000,100000]
plt.figure(figsize=(12,6))
for t in T:
    plt.plot(e_mu,n_bar(t),label='T={0}'.format(t))
plt.xlabel('$\epsilon-\mu$',fontsize=17)
plt.ylabel('$\overline{n}$',fontsize=17)
plt.grid()
plt.ylim(ymax=1.05)
plt.legend()
plt.show()
```
