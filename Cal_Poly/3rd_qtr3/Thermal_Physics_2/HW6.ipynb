```python
from scipy.optimize import brentq
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
h=6.626e-34
c=3e8
k=1.38e-23
```

###7.37

```python
print brentq(lambda x: 3*(1-np.exp(-x))-x,2,3)
```

###7.38

```python
x=np.linspace(0,1.6e-18,10000)
y=lambda x,T: 8*np.pi/h/c*x**3/(np.exp(x/k/T)-1)
plt.figure(figsize=(10,5))
plt.plot(x,y(x,3000),label='T=3000')
plt.plot(x,y(x,6000),label='T=6000')
plt.xlabel('$\epsilon$ $(eV)$',fontsize=15)
plt.ylabel('$U(\epsilon)$',fontsize=15)
plt.legend()
plt.show()
```
