Finding Possible Detecor Combinations
-

```python
#in the multilateration equations provided by wikipedia each calculation requires a set of five detecors which are put into three simple plane equations. two detectors are selected; the location of the first detector
#represents the location of the origin, the other detecor is determined to be a 'camparison detector'. these two detecors are used in all three equations. the remaining detecors are used to make
#the three resulting equations unique.

#so in order to have three unique equations the three detecors which make each equation distinct must be different.
#the following code finds all combinations of three detectors such that each combination contains three different codetecors where each detector is labeled using a number.

total_list = []
for i in range(len(DetectorIn)):
    for j in range(len(DetectorIn)):
        for k in range(len(DetectorIn)):
            total_list.append((i+1)**2+(j+1)**2+(k+1)**2)
            
from collections import defaultdict
            
tally = defaultdict(list)
for i,item in enumerate(total_list):
    tally[item].append(i)
    
comb_list = []
for i in range(len(DetectorIn)):
    for j in range(len(DetectorIn)):
        for k in range(len(DetectorIn)):
            comb_list.append([i+1,j+1,k+1])

comb_reduced0 = []
for x in tally:
    comb_reduced0.append(comb_list[tally[x][0]])

repeat_comb = []
for i in comb_reduced0:
    tally2 = defaultdict(list)
    for i,item in enumerate(i):
        tally2[item].append(i)
    repeat_comb.append(tally2)

comb_reduced = []
for i in range(len(repeat_comb)):
    if len(repeat_comb[i])==3:
        comb_reduced.append(comb_reduced0[i])
        
print comb_reduced
```

```python
# all code in this cell originates from: http://rosettacode.org/wiki/Reduced_row_echelon_form#Python
# this code perfomrs a reduced row echelon operation on a 3x4 matrix.

def ToReducedRowEchelonForm( M):
    if not M: return
    lead = 0
    rowCount = len(M)
    columnCount = len(M[0])
    for r in range(rowCount):
        if lead >= columnCount:
            return
        i = r
        while M[i][lead] == 0:
            i += 1
            if i == rowCount:
                i = r
                lead += 1
                if columnCount == lead:
                    return
        M[i],M[r] = M[r],M[i]
        lv = M[r][lead]
        M[r] = [ mrx / lv for mrx in M[r]]
        for i in range(rowCount):
            if i != r:
                lv = M[i][lead]
                M[i] = [ iv - lv*rv for rv,iv in zip(M[r],M[i])]
        lead += 1
 
 
mtx = [
       [1,2,0,4],
       [2,1,0,5],
       [3,1,0,6]]
 
ToReducedRowEchelonForm( mtx )
 
for rw in mtx:
  print ', '.join( (str(rv) for rv in rw) )
```
