```python
def random_xyzthetaphi(N):
    '''creates a set of x,y, and z values which lie within a cylindrical region which has a radius of '1.3' and a height of '.2'. the region is also centered at a height of 15km.
    in addition to position it includes a set of incidence angles and energies which are associated with each particle.'''
    from random import gauss
    global x_list
    global y_list
    global z_list
    global theta_list
    global phi_list
    x_list = [10]*N 
    y_list = [10]*N
    z_list = [0]*N
    theta_list = [0]*N
    phi_list = [0]*N
    for i in range(N):
        #phi is measured as the angle from the line perpedicular to the ground. 0 degrees is directed towards or into the ground.
        #for the purpose of efficiency phi is limited to being less than 0.2*pi.
        phi_list[i] = 0*pi*ranf()
        # theta is measured as the angle on the plane parrallel to the ground. 0 degrees is North
        theta_list[i] = 2*pi*ranf()
        # z is the height at which the first interaction occures. (all interactions occure within 200 meters of a height of 15km.
        z_list[i] = 14.9+.2*ranf()
        #the x and y cordinates of each initial interaction are limited to the inside of a circle whose radius is 1.3km.
        x_list[i] = 2.6*ranf()-1.3
        while -sqrt(1.3**2-x_list[i]**2)>y_list[i] or y_list[i]>sqrt(1.3**2-x_list[i]**2):
            y_list[i] = 2.6*ranf()-1.3
    return
```

```python
# Google Elevation Data For Detector Positions
D1 = [0.0,0.0,0]
D2 = [0.2,0.2,0]
D3 = [-0.2,-0.2,0]
D4 = [0.2,-0.2,0]
D5 = [-0.2,0.2,0]
D6 = [0.4,0.4,0]
D7 = [-0.4,-0.4,0]
D8 = [0.4,-0.4,0]
D9 = [-0.4,0.4,0]

DA = [D1,D2,D3,D4,D5,D6,D7,D8,D9]
# Saving detecotr locations
savetxt('detector_locations.dat',DA,fmt='%f')

def rmv_interactions(N):
    from random import gauss
    from itertools import ifilter
    from random import uniform
    global dtimes_alledit
    global datatimes_all
    global energylist_all
    datatimes_all = []
    energylist_all = []
    dtimes_all = []
    C = 3*10**5
    Num_Good_Interactions = 0  
    print '--------------------------------------------------------------------------------------------------------------------------------------------------------------------------------'
    for i in range(N):
        print
        # defining variables: xi = x cordinate of a point of first interaction, yi = y cordinate of a point of first interaction,
        # z = z cordinate of a point of first interaction, theta = angle of the resulting cone  where 0 degrees is taken to be north,
        # phi = the angle of the resulting cone relative to a line perpendicular to the ground
        xi = x_list[i]
        yi = y_list[i]
        z = z_list[i]
        theta = theta_list[i]
        phi = phi_list[i]
        
        # x and y cordinates of the resulting ellipse relative to the point of first interaction
        xofEint = tan(phi)*z*cos(theta)
        yofEint = tan(phi)*z*sin(theta)
    
        # l = long axis of the ellipse, r = short axis of elipse
        l = z*(tan(phi+2*pi/180.)-tan(phi))
        r = z*tan(pi/180)/cos(phi)
     
        # x and y cordinates of the ellipse relative to the origin, where xi,yi are cordinates of the initial interaction,
        # and xofEint,yofEint are cordinates of the elipse relative to the point of interaction.
        xofEorig = xi+xofEint
        yofEorig = yi+yofEint
        
        # equations for rotating a cartisian cordinate system by an angle theta. X and Y cordinates x and y are transformed into the rotated X' and Y' cordinate system x_rotated and y_rotated.
        # x_rotated = x*cos(theta/2.)+sin(theta/2.)*y
        # y_rotated = -x*sin(theta/2.)+cos(theta/2)*y
        
        # altering dectecor positions so their x and y cordinates are relative to the major and minor axis of the ellipse respectively
        for a in range(len(DA)):
            DA[a][0] = (DA[a][0]-xofEorig)*cos(theta)+sin(theta)*(DA[a][1]-yofEorig)
            DA[a][1] = -(DA[a][0]-xofEorig)*sin(theta)+cos(theta)*(DA[a][1]-yofEorig)
     
        # upper and lower bounds on the y axis of the elipse. the equation 0 <= 1-((DA[j][0])**2 represents the upper and lower bounds on the x axis.
        posyE = lambda j: sqrt(r*(1-((DA[j][0])**2)/l))+yofEorig
        negyE = lambda j: -sqrt(r*(1-((DA[j][0])**2)/l))+yofEorig
 
        # determining whether detector is within the elipse.
        counter = 0
        for b in range(len(DA)):
            print posyE(b),',',DA[b][1],',',negyE(b)
            print posyE(b)>DA[b][1],',',negyE(b)<DA[b][1]
            # plugging in x positions of the detecors into the equations to see if the y values lie within the bounds of the elipse.
            if posyE(b)>DA[b][1] and negyE(b)<DA[b][1]:
                counter = counter + 1
        print
        print 'number of detectors hit:',counter
        print '......................................';print
        
#.......................................................................................................................................................................................................................        
        
        # If 5 or more detecotors are hit then the time that it took for each muon to reach one of those detectors is calculated
        dtimes = ['not valid']
        datatimes = []
        ddatatimes = []
        energyi_list = []
        energyi_print = []
        if counter==5 or counter>5:
            print 'xyz cordinates of valid interaction:',[xi,yi,z]
            print;print '-----------------------------------------------------------------------------------------------------------------------------------';print
            # Finding the times that it took each muon to travel to the detector.
            times = []
            for d in range(len(DA)):
                if posyE(d)>DA[d][1] and negyE(d)<DA[d][1]:
                    E = gauss(4,1.5)
                    title = 'D%s' %(d+1)
                    times.append(title)
                    times.append(sqrt((xi-DA[d][0])**2+(yi-DA[d][1])**2+DA[d][2]**2)/(C*sqrt(1-((.1057)**2)/E**2)))
                    datatimes.append(sqrt((xi-DA[d][0])**2+(yi-DA[d][1])**2+DA[d][2]**2)/(C*sqrt(1-((.1057)**2)/E**2)))
                    energyi_list.append(E)
                else:
                    datatimes.append('X')
                    energyi_list.append('X')
            # making those times into differences in times where the fastest time to the detector is at zero seconds.
            dtimes = []
            for c in range(len(datatimes)):
                if datatimes[c]!='X':
                    ddatatimes.append(datatimes[c]-min(datatimes))
                else:
                    ddatatimes.append('X')
            for f in range(len(datatimes)):
                if ddatatimes[f] != 'X':
                    ddatatimes[f] = ddatatimes[f]+ddatatimes[f]*uniform(-0.05,0.05)
                
            for e in range(len(times)):
                if e%2 != 0:
                    dtimes.append(times[e-1])
                    dtimes.append(times[e]-min(times))
            for e in energyi_list:
                if e!='X':
                    Etitle = 'D%s' %(energyi_list.index(e)+1)
                    energyi_print.append(Etitle)
                    energyi_print.append(e)
                    
            # Times of each muon detection at D1, D2, D3, etc. respectively, with the first detection at t = 0.
            print 'Times:', times
            print 'Energie:', energyi_print
            print;print '--------------------------------------------------------------------------------------------------------------------------------------------------------------------------------'
            Num_Good_Interactions += 1

            print '--------------------------------------------------------------------------------------------------------------------------------------------------------------------------------'
            datatimes_all.append(ddatatimes)
            energylist_all.append(energyi_list)
            

        
        dtimes_all.append(dtimes)
        
#------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------
#------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------
#------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------ 
        
    print
    dtimes_alledit = []
    def check_item(s):
        return s != ['not valid']
    for g in ifilter(check_item, dtimes_all):
        dtimes_alledit.append(g)


    print
    print '.......................................................................................................'
    print 'NUMBER OF ANALYZABLE INTERACTIONS:', Num_Good_Interactions
    print '.......................................................................................................'

    print 'TIMES FOR ALL VALID INTERACTIONS:'
    print dtimes_alledit

    print '.......................................................................................................'
    print
    print 'NUMBER OF ANALYZABLE INTERACTIONS:',Num_Good_Interactions
    print '.......................................................................................................'
    print
    print 'ACTUAL DATA:';print
    print '____t____'
    print datatimes_all; print
    print'____E____'
    print energylist_all
    return



#>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>
#>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>
#>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>



N = 20
random_xyzthetaphi(N)
# plots first interaction positions and 
from matplotlib import pyplot
import pylab
from mpl_toolkits.mplot3d import Axes3D
fig = pylab.figure()
ax = Axes3D(fig)
ax.scatter(x_list,y_list,z_list)

Dx_list = []
Dy_list = []
Dz_list = []

for i in range(len(DA)):
    Dx_list.append(DA[i][0])
    Dy_list.append(DA[i][1])
    Dz_list.append(DA[i][2])
print Dx_list
print Dy_list
ax.scatter(Dx_list,Dy_list,Dz_list, color='r')
pyplot.show()
figure(figsize = (7,7))
plot(x_list,y_list, 'b.')
plot(Dx_list,Dy_list, 'r+', markersize=10)
show()
rmv_interactions(N)
```
