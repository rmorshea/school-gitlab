```python
N = 10000
a = 1.5
B = 6
f = lambda x: (a*B**(-a))*x**(a-1)*exp(-(x/B)**a)
fin = lambda x: B*(-log(1-x))**(1/a)
CDF = lambda x: 1-exp(-(x/B)**a)

x = arange(0,30.001,0.001)
z = ranf(N)
xsamples = fin(z)

plot(x,CDF(x), label='Cumulative distribution')
plot(x,f(x), linewidth = 1.5, label='True distribution')
hist(xsamples,50,normed=True, label='Sampled distribution')
legend(loc='center right')
show()
```

```python
N = 100000
xmin = 0
xmax = 1
ymin = 0
ymax = 2.076

g = lambda x: 60*(x**3)*(1-x)**2
x = arange(0,1.001,.001)
CDF = cumsum(g(x))/sum(f(x))

samples_x = xmin + (xmax-xmin)*ranf(N)
samples_y = ymin +(ymax-ymin)*ranf(N)
mask = (samples_y<g(samples_x))

plot(x,CDF, linewidth = 1.75, label='Cumulative distribution')
plot(x,g(x), linewidth = 1.75, label='True distribution')
hist(samples_x[mask],50,normed=True, label='Sampled distribution')
ylim(0,3.2)
legend(loc='upper left')
show()
```
