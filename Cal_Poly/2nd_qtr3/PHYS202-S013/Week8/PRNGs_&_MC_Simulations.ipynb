```python
x = arange(0.,pi+0.001,0.01)
f = lambda x: sin(x)/2.
F = lambda x: (cos(0)-cos(x))/2.

plot(x,f(x),label='True distribution',lw = 3)
plot(x,F(x),label='Cumulative distribution',lw = 3)

xlim(0,3.5)
ylim(0,1.1)
legend(loc=0)
show()
```

```python
def distribute1D(x,prob,N):
    z = numpy.random.random_sample(N)
    G = cumsum(prob)
    b = sum(prob)
    Q = G/b
    T = Q.searchsorted(z)
    new = x[T]
    return new
```

```python
C = 1
f = lambda x: C*exp(-C*x)
x = arange(0,5.01,0.01)
hist(distribute1D(x,f(x),10000),50,normed=True,label='Sampled distribution')
```
