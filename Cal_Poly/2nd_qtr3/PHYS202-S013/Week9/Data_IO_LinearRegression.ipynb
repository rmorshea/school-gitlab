```python
def LinearLeastSquaresFit(x,y):
    '''Take in arrays representing (x,y) values for a set of linearly varying data and perform a linear least squares regression. Returns the resulting slope and intercept parameters of the best fit line with their uncertainties.'''
    import numpy
    import math
    n = float(len(x))
    xbar = mean(x)
    ybar = mean(y)
    xx = mean(x**2)
    xy = mean(x*y)
    
    m = (xy - (xbar*ybar))/(xx - (xbar**2))
    b = ((xx*ybar) - (xbar*xy))/(xx - (xbar**2))
    
    delta = y-(m*x + b)
    d = mean(delta**2)
    
    sig_m = sqrt((1/(n-2))*(d/(xx - (xbar**2))))
    sig_b = sqrt((1/(n-2))*((d*xx)/(xx - (xbar**2))))

    
    return m,b,sig_m,sig_b
```

```python
x,y = loadtxt('linearData.dat', unpack=True, usecols=[0,1])
m,b,sig_m,sig_b = LinearLeastSquaresFit(x,y)

print x,y
```

```python
plot(x,m*x+b, label='y = %.4f x + %.4f'%(m,b), lw=2)
plot(x,(m+sig_m)*x + (b+sig_b),'r--', label = 'y = (%.4f + %.4f) x + (%.4f + %.4f)'%(m,sig_m,b,sig_b), lw=1)
plot(x,(m-sig_m)*x + (b-sig_b),'g--', label = 'y = (%.4f - %.4f) x + (%.4f - %.4f)'%(m,sig_m,b,sig_b), lw=1)
legend(loc = 'upper right',prop={'size':8})
title("Linear Approximation of Position Data", size=13)
xlabel("X",size = 10)
ylabel("Y",size = 10)
show()
```
