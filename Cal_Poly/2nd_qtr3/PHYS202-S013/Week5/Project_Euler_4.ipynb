```python
from numpy import *

def plndrm_finder(digits):
    plndrms = []
    for i in range(1,10**digits):
        for j in range(i,10**digits):
            num = str(i*j)
            numr = num[::-1]
            if num == numr:
                plndrms.append(i*j)
    return plndrms

print 'All Two Digit Product Palindromes:'
print
print plndrm_finder(2)
print '--------------------------------------------------------------------------------------------------------------------------------------------'
print
print 'Max Palindrome Of Two Digit Product =', max(plndrm_finder(2))
print
print '--------------------------------------------------------------------------------------------------------------------------------------------'
print
print 'All Three Digit Product Palindromes:'
print
print plndrm_finder(3)
print '--------------------------------------------------------------------------------------------------------------------------------------------'
print
print 'Max Palindrome Of Three Digit Product =', max(plndrm_finder(3))
          
```
