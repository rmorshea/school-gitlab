```python
import scipy.integrate as integrate
x = arange(0,3,.1)
y = x
y = 0
y = e**(-(x**2))
Y=zeros(x.shape,float)
Y[0]=0
Y[1:]=integrate.cumtrapz(y,x,.01,)
plot(x,Y)
xlabel('X')
ylabel('$e^{-x^2}$')
title('Integral of $e^{-x^2}$')
show()
```

```python
import scipy.integrate as integrate
n = 1 #ground state
L = 1
A = sqrt(2/L)
k = (n*pi)/L
psi = lambda x: A*sin(k*x)

N = 10000000
a = L/3.
b = L/2.

x = arange(a,b+0.01, ((b-a)/N))
psi_new = lambda X: (psi(X))**2
psi_trap = integrate.trapz(psi_new(x),x)

psi_quad = integrate.quad(psi,a,b)

x_rang = arange(0,L+0.01,0.01)


plot(x_rang,(psi(x_rang))**2, )

title('a particle in a one-dimensional box')
xlabel('X')
ylabel('$|Psi|^2$, n=1')
show()

print
print "Trapazoidal =",psi_trap
print "Required 1E7 bins"
print "Quadratic =",psi_quad[0]
```

```python
def H(x,n):
    d = [1,2*x]
    for i in range(1,n+1):
        d.append(2*x*d[i]-2*i*d[i-1])
    return d[n]

psi = lambda x,n: 1./(sqrt((2.**n)*(math.factorial(n))*sqrt(pi)))*exp((-x**2.)/2.)*H(x,n)

X = arange(-4,4.1,.1)
for i in range(0,4):
    plot(X,psi(X,i), label = '$\psi_{%s}(x)$' %i)
ylabel('$\psi_{n}(x)$', fontsize = 16)
xlabel('X')
xlim(xmax = 4, xmin = -4)
ylim(ymax = 1.5, ymin = -1.5)
legend(loc = 'lower right')
title('Harmonic oscillator in nth energy level')
show()
```

```python
mass_density = lambda y,z: 100*((y+0.1)+z**2)
l = 0.2
w = 0.2
h = 1.0
N = 1e6
ri = lambda i,y,z: i*mass_density(y,z)

Ii = lambda i,y,z: ((i-ri(i,y,z))**2)*mass_density(y,z)
x = l*random_sample(N); y = w*random_sample(N); z = h*random_sample(N)
totalMass = (l*w*h/N)*sum(mass_density(y,z))
print "total mass:",totalMass
print ""

rx = (l*w*h/N)*sum(ri(x,y,z))
ry = (l*w*h/N)*sum(ri(y,y,z))
rz = (l*w*h/N)*sum(ri(z,y,z))
print "r_x:", rx
print "r_y:", ry
print 'r_z:', rz
print ""
Ix = (l*w*h/N)*sum(Ii(x,y,z))
Iy = (l*w*h/N)*sum(Ii(y,y,z))
Iz = (l*w*h/N)*sum(Ii(z,y,z))
print "I_x:", Ix
print "I_y:", Iy
print 'I_z:', Iz
```
