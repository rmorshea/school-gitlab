Publication ready plot template
====
Simple graphing template for matplotlib functions and data plots.

Line Plots
-

```python
from numpy import *
y = arange(-5,5.2,.2)
#creates a range of inputs to the functions below
E = cosh(y) #creates a function that has an input of the array 'y'
p = sinh(y) #creates a function that has an input of the array 'y'
```

```python
plot(y,E, 'r', color='red', label = '$cosh(y)$')   #controls what is graphed, the type of line, color of the line, and the lable associated with the line
plot(y,p, 'r:', color='blue', label = '$sinh(y)$') #controls what is graphed, the type of line, color of the line, and the lable associated with the line
ax = gca()
ax.spines['left'].set_position('center')   #adjusts the postions of the y-axis(or spine) of the graph
ax.spines['right'].set_color('none')       #removes the spines around the graph
ax.spines['bottom'].set_position('center') #adjusts the postions of the x-axis(or spine) of the graph
ax.spines['top'].set_color('none')         #removes the spines around the graph
annotate('x-axis', xy=(0,0), xytext=(6.4,0), fontsize = 10)  # normal x-axis label doesn't work in this context
annotate('y-axis', xy=(0,0), xytext=(-.6,-90), fontsize = 10)  # normal y-axis label doesn't work in this context
title('My Plot', fontsize = 14)  # plot title
xticks(arange(-2*pi, 2*pi+1, pi/2),['$-2{\pi}$','$-3{\pi}/2$','$-{\pi}$','$-{\pi}/2$','$0$', '$+{\pi}/2$', '$+{\pi}$','$3{\pi}/2$', '$2{\pi}$'])
#creates the x-axis labels or -2pi, -3pi/2, etc...
legend(loc = 'lower right') #shows legend
show() #makes graph presentable
```
