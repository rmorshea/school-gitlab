Plotting Electric Potentials
=

$V(x,y) = \frac{kq}{\sqrt{x^2+(y-d)^2}} - \frac{kq}{\sqrt{x^2+(y+d)^2}}$

```python
k = 1 ; q = 1 ; d = 1
xmin = -4. ; xmax = 4.
ymin = -4. ; ymax = 4.
xdivs = 201. ; ydivs = 201.
dx = (xmax-xmin)/xdivs ; dy = (ymax-ymin)/ydivs
x = arange(xmin,xmax,dx)
y = arange(xmin,ymax,dy)
print len(x),len(y)
```

```python
X,Y = meshgrid(x,y)
Vd = (k*q)/((X**2+(Y-d)**2))**(1/2.) + (k*(-q))/((X**2+(Y+d)**2))**(1/2.)
#dipole potential
Vq = (k*q)/((X**2+(Y-d)**2))**(1/2.) + (k*(q))/((X**2+(Y+d)**2))**(1/2.) + (k*(-q))/(((X-d)**2+(Y)**2))**(1/2.) + (k*(-q))/(((X+d)**2+(Y)**2))**(1/2.)
#quadrupole potential

"""'dx' and 'dy' are chosen as such in order to avoid pluging in numbers that will land on the position of the poitn charges
in this case the positions of the charges are at -1,1 based on the value of 'd'
if we choose 'd' to be divisible by 'dx' and 'dy' then the computer will have a difficult time representing a number as large as infinity"""

print X.ndim; print Y.ndim; print; print X; print; print Y; print; print Vd
```

```python
from numpy import *
from mpl_toolkits.mplot3d import Axes3D
fig = figure()
k=1; q=1; d=1
ax = fig.gca(projection = '3d')
p = ax.plot_surface(X,Y,Vd, rstride = 1, cstride = 1, lw = 0, cmap=cm.Set1_r)
```

```python
fig = figure()
k=1; q=1; d=1
ax = fig.gca(projection = '3d')
p = ax.plot_surface(X,Y,Vq, rstride = 1, cstride = 1, lw = 0, cmap=cm.Set1_r)
#help(Axes3D.plot_surface)
```

```python
#defining easily accesible functions for point potentials and dipoles
def PointPotential(k,x,y,q,posx,posy):
    """returns the electric potential due to the charge on point particle q"""
    Vp = (k*q)/(((X-posq_x)**2+(Y-posq_y)**2))**(1/2.)
    return Vp
def DipolePotential(k,x,y,q1,q2,posq1_x,posq1_y,posq2_x,posq2_y):
    '''returns the electric potential due to the charge on the particles q1 and q2'''
    Vd = (k*q1)/(((X-posq1_x)**2+(Y-posq1_y)**2))**(1/2.) + (k*q2)/(((X-posq2_x)**2+(Y-posq2_y)**2))**(1/2.)
    return Vd
```

```python
k = 1
q = 1
posq_x = 0
posq_y = 0
xmin = -.5 ; xmax = .5
ymin = -.5 ; ymax = .5

xdivs = 51. ; ydivs = 51.
dx = (xmax-xmin)/xdivs ; dy = (ymax-ymin)/ydivs
x = arange(xmin,xmax,dx)
y = arange(xmin,ymax,dy)
X,Y = meshgrid(x,y)

V_Func1 = PointPotential(k,X,Y,q,posq_x,posq_y)


k = 1
q1 = 1 ; q2 = -1
posq1_x = .25 ; posq1_y = .25
posq2_x = -.25 ; posq2_y = -.25

V_Func2 = DipolePotential(k,X,Y,q1,q2,posq1_x,posq1_y,posq2_x,posq2_y)
```

```python
fig = figure()
ax = fig.gca(projection = '3d')
p = ax.plot_surface(X,Y,V_Func1, rstride = 1, cstride = 1, lw = 0, cmap=cm.Set1)


fig = figure()
ax = fig.gca(projection = '3d')
p2 = ax.plot_surface(X,Y,V_Func2, rstride = 1, cstride = 1, lw = 0, cmap=cm.Set1)
```
