```python
x = arange(0,14.01,0.01)
y = m*x+b

figure(figsize=(8,5))
plot(x,m*x+b, label='y = %.4f x + %.4f'%(m,b), lw=2)
plot(x,(m+sig_m)*x + (b+sig_b),'r--', label = 'y = (%.4f + %.4f) x + (%.4f + %.4f)'%(m,sig_m,b,sig_b), lw=1)
plot(x,(m-sig_m)*x + (b-sig_b),'b--', label = 'y = (%.4f - %.4f) x + (%.4f - %.4f)'%(m,sig_m,b,sig_b), lw=1)
legend(loc = 'upper right',prop={'size':8})
title("Linear Approximation of Position Data", size=13)
xlabel("t",size = 10)
ylabel("N",size = 10)
show()
```

```python
waveform_2 = load('waveform_2.npy')
t = np.arange(len(waveform_2))
plot(t,waveform_2)
xlabel('time (ns)')
ylabel('intensity (bins)')
show()
```

```python
def fitfunc(t,a,b,c,d,mu0,sigma0,mu1,sigma1,mu2,sigma2):
    return a+b*exp(-((t-mu0)/sigma0)**2)+c*exp(-((t-mu1)/sigma1)**2)+d*exp(-((t-mu2)/sigma2)**2)
```

```python
x0 = array([3.,25.,13.,8.,16.,1.,22.,1.,27.,1.])
```

```python
from scipy.optimize import curve_fit
popt,pcov = curve_fit(fitfunc,t,waveform_2,x0)
plot(t,waveform_2,label='data')
plot(t,fitfunc(t,popt[0],popt[1],popt[2],popt[3],popt[4],popt[5],popt[6],popt[7],popt[8],popt[9]),label='fit',linewidth=2)
show()
```
