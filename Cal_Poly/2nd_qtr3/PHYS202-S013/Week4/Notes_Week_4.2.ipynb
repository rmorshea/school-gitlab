Use Boolean Masks For Finding Data
=

```python
from numpy import *
theta = arange(0.,2*pi,pi/100)
plot(theta,sin(theta))
show()
```

```python
mask = (sin(theta)>0) & (sin(theta)<0.5)
print mask
plot(theta,sin(theta), theta[mask], sin(theta[mask]))
```

```python
y = sin(theta[mask])
jj = (y == max(y)); print jj
y[jj] = NaN
plot(theta,sin(theta))
plot(theta[mask],y,linewidth = 4)
```

Interpolation
=

```python
theta = arange(0,2*pi+0.01, pi/10)
mask = (theta == .5)
print mask
```

what we want to be able to do is numerically interpolate between two values so that even though our $d\theta$ doesnt return a value where theta == .5 we can still recieve a value.

```python
newTheta = 0.5; newSin = 0.
from scipy.interpolate import interp1d
#help(interp1d)
sinInterp = interp1d(theta,sin(theta))
newSin = sinInterp(newTheta)

print newSin
plot(theta,sin(theta),newTheta,newSin, 'r+', markersize = 10, markeredgewidth = 2)
show()

#note that the marker isn't exactly on the line. this is fine it's simply because python had to linearly interpolate it.
```

```python
thetaNew = arange(0,2*pi+0.1,pi/30)
sinNew = interp1d(theta,sin(theta))(thetaNew)
plot(theta,sin(theta),thetaNew,sinNew,'r.')
```

```python
#so how good is the approximation? subtract the two to see

plot(thetaNew,sinNew-sin(thetaNew))
show()
```

Note that the approximation is worst at the peaks of the funtions where $\frac{d^2(sinNew)}{d(thetaNew)^2}$ $ $ is greatest.
