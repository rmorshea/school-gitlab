```python
def finiteDifference(x,y):
    """takes as arguments arrays of values x and y(x) and returns the derivative of y with respect to x"""
    dydx = zeros(y.shape, float)
    dydx[0:-1] = diff(y)/diff(x)
    dydx[-1] = (y[-1] - y[-2])/(x[-1] - x[-2])
    return dydx

x = arange(0,10.1,.1)
y = x**5
actual_dydx = 5*x**4

percent_error = (actual_dydx-finiteDifference(x,y))/actual_dydx

print 'DIFFERENCE:'
print actual_dydx - finiteDifference(x,y)

plot(x,percent_error, label = 'percent error')
xlabel('X', fontsize = 9)
ylabel('Y', fontsize = 9)
title('Centered Finite Difference Error', fontsize = 12)
legend(loc = 'lower right')
show()
```

```python
h = .1
x = arange(0,10,h)
y = x**5

def fourPtFiniteDiff(x,y):
    """computes the derivative using the higher order method called the four-point center differencing method"""
    dydx = zeros(y.shape, float)
    for i in range(2,len(y)-2):
        dydx[i] = (y[i-2]-8*y[i-1]+8*y[i+1]-y[i+2])/(12*h)
    dydx[-1] = (y[-1]-y[-2])/(x[-1]-x[-2])
    dydx[-2] = (y[-2]-y[-3])/(x[-2]-x[-3])
    dydx[0] = (y[1]-y[0])/(x[1]-x[0])
    dydx[1] = (y[2]-y[1])/(x[2]-x[1])
    return dydx

plot(x,fourPtFiniteDiff(x,y),linewidth=2.0, label = '4t-centered derivative')
plot(x,5*x**4,'r.', label = 'analytical function')
xlabel('X', fontsize = 9)
ylabel('Noisy Data and Derivatives', fontsize = 9)
title('4pt-Derivative Comparison', fontsize = 12)
legend(loc = 'upper left')
show()
```
