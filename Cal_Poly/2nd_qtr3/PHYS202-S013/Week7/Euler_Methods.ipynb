Example 1

```python
a = -9.81
t = arange(0,10,0.01)
v = zeros(t.shape)
dt = 0.01
for i in range(t.shape[0]):
    v[i]=v[i-1]+a*dt
```

```python
plot(t,v)
xlabel('t')
ylabel('v(t)')
show()
```

```python
r = zeros(t.shape)
for i in range(t.shape[0]):
    r[i]=r[i-1]+v[i-1]*dt
```

```python
plot(t,r)
xlabel('t')
ylabel('r(t)')
show()
```

```python
dx = 0.25
x = arange(0,10,dx)
y = zeros(x.shape)
y2 = zeros(x.shape)
for i in range(x.shape[0]):
    y[i] = y[i-1]+(x[i]+2*y[i-1])*dx
for i in range(x.shape[0]):
    y2[i] = 0.25*exp(2*x[i])-0.5*x[i]-0.25
```

```python
figure(figsize=(7,7))
plot(x,y, label='$y(x)_{0.25}$')
xlabel('x')
ylabel('y(x)')
dx = 0.02
plot(x,y, 'r.', label='$y(x)_{0.02}$')
plot(x,y2, label='$y(x)_{actual}$')
legend(loc='upper left')

show()
```
