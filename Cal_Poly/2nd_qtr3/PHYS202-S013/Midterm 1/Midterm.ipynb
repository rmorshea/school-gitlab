```python
def is_anagram(s1,s2):
    counter = 0
    if len(s1) == len(s2):
        for i in range(0,len(s1)):
            for j in range(0,len(s1)):
                if s1[i] == s2[j]:
                    counter += 1
        if counter == len(s1):
            return True
        else:
            return False
        
    else:
        return False

print 'TRUE CASE:'
print 'is_anagram returns:', is_anagram('tam','mat')

print

print 'FALSE CASE:'
print 'is_anagram returns:', is_anagram('tam','mweet')
print 'is_anagram returns:', is_anagram('tam','mwt')
```
