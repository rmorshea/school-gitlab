```python
import numpy as np
import math
from math import factorial as fact
import csv
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
T_rot = range(0,500,2)
T_rot.remove(0)
T_rot.insert(0,.1)
for i in range(len(T_rot)):
    T_rot[i]=float(T_rot[i])
T_rot=np.array(T_rot)

#print T_rot
```

```python
plt.figure(num=None, figsize=(8, 6), dpi=80, facecolor='w', edgecolor='k')
def Z_rot(j,T): #input starting j and go up in integer steps to j+10
    b=1./(T*8.617e-5) #boltzmann const in ev/K
    erot=0.0076 #eV this is for diatomic hydrogen
    a=(2.*j+1.)*np.exp(-b*(j*(j+1.))*erot)+(2.*(j+1.)+1.)*np.exp(-b*((j+1.)*((j+1.)+1.))*erot)+(2.*(j+2.)+1.)*np.exp(-b*((j+2.)*((j+2.)+1.))*erot)+(2.*(j+3.)+1.)*np.exp(-b*((j+3.)*((j+3.)+1.))*erot)+(2.*(j+4.)+1.)*np.exp(-b*((j+4.)*((j+4.)+1.))*erot)+(2.*(j+5.)+1.)*np.exp(-b*((j+5.)*((j+5.)+1.))*erot)+(2.*(j+6.)+1.)*np.exp(-b*((j+6.)*((j+6.)+1.))*erot)+(2.*(j+7)+1.)*np.exp(-b*((j+7.)*((j+7.)+1.))*erot)+(2.*(j+8.)+1.)*np.exp(-b*((j+8.)*((j+8.)+1.))*erot)+(2.*(j+9.)+1.)*np.exp(-b*((j+9.)*((j+9.)+1.))*erot)+(2.*(j+10.)+1.)*np.exp(-b*((j+10.)*((j+10.)+1.))*erot)
    return a
plt.plot(T_rot,Z_rot(0.,T_rot))
plt.xlim(xmax=5000)

plt.show()
```

```python
plt.figure(num=None, figsize=(8, 6), dpi=80, facecolor='w', edgecolor='k')

def ln_Z(T):
    return np.log(Z_rot(0,T))

def U_over_N(T):
    h=1
    a=(ln_Z(T+h/2.)-ln_Z(T-h/2.))/h
    return a*(T**2.)*8.617e-5

plt.xlim(0,5000)
plt.xlabel("$T$ (K)",fontsize=15)
plt.ylabel("$U/N$ (eV)", fontsize=15)
plt.plot(T_rot,U_over_N(T_rot))
plt.show()
```

```python
def C_over_Nk(T):
    h=.1e-2
    a=(U_over_N(T+h/2.)-U_over_N(T-h/2.))/h
    return a/8.617e-5
```
