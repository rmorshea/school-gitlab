The goal of this final computer assignment is to predict the molar specific heat of H2 molecules, 
as seen experimentally in Figure 1.13 in the textbook. Hydrogen molecules have translational 
energy, rotational energy, and vibrational energy. All degrees of freedom play a role.

```python
import numpy as np
import math
from math import factorial as fact
import csv
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
k=8.617e-5
T=np.linspace(1,10000,1000)
lnT=np.linspace(0,10000,1000)
e=0.53
E_ha=[i*e for i in range(11)]
E_an=[1.03*i-0.03*i**2 for i in range(11)]
Z_vib=[sum([math.exp(-E/float(k*t)) for E in E_an]) for t in T]
lnZ_vib=np.log(np.array(Z_vib))

print E_ha
print E_an
```

```python
plt.figure(figsize=(10,5))
plt.plot(T,Z_vib)
plt.xlabel('$T$ (K)',fontsize=14)
plt.ylabel('$Z$',fontsize=14)
plt.xlim(0,5100)
plt.title('$Z$ vs $T$')

plt.show()
plt.figure(figsize=(10,5))
plt.plot(lnT,lnZ_vib)
plt.xlabel('$T$ (K)',fontsize=14)
plt.ylabel('$ln(Z)$',fontsize=14)
plt.xlim(0,5000)
plt.ylim(0,1.1)
plt.title('$ln(Z)$ vs $T$',fontsize=14)
plt.show()
```

```python
def deriv(X,Y):
    '''Derivative of X with respect to Y'''
    if len(X)!=len(Y):
        raise ValueError('X and Y must have same first dimension')
    D=[]
    for i in range(len(X)):
        if i==len(X)-1:
            break
        if i!=0:
            D.append(float(X[i+1]-X[i-1])/float(Y[i+1]-Y[i-1]))
    return D

def find_U(Y,X):
    '''Derivative of X with respect to Y'''
    if len(X)!=len(Y):
        raise ValueError('X and Y must have same first dimension')
    D=[]
    for i in range(len(Y)):
        if i==len(Y)-1:
            break
        if i!=0:
            D.append(k*X[i]**2*float(Y[i+1]-Y[i-1])/float(X[i+1]-X[i-1]))
    return D
```

```python
U=find_U(lnZ_vib,lnT)
plt.figure(figsize=(10,5))
plt.plot(lnT[1:-1],U)
plt.xlabel('$T$ (K)',fontsize=13)
plt.ylabel('$U$ (eV$/N$)',fontsize=13)
plt.title('$U$ vs $T$',fontsize=13)
plt.show()
```

```python
def find_Cv(X,Y):
    '''Derivative of X with respect to Y'''
    if len(X)!=len(Y):
        raise ValueError('X and Y must have same first dimension')
    D=[]
    for i in range(len(X)):
        if i==len(X)-1:
            break
        if i!=0:
            D.append(float(X[i+1]-X[i-1])/float(Y[i+1]-Y[i-1])/k)
    return D
```

```python
Cv=find_Cv(U,lnT[1:-1])
plt.figure(figsize=(10,5))
plt.plot(lnT[2:-2],Cv)
plt.xlabel('$T$ (K)',fontsize=13)
plt.ylabel('$C_v$',fontsize=13)
plt.title('$C_v$ vs $T$',fontsize=13)
plt.show()
print len(E_an)-len(T)
```

```python
T_rot = range(0,5002,2)
T_rot.remove(0)
T_rot.insert(0,.1)
for i in range(len(T_rot)):
    T_rot[i]=float(T_rot[i])
T_rot=np.array(T_rot)

#print T_rot
```

```python
plt.figure(num=None, figsize=(8, 6), dpi=80, facecolor='w', edgecolor='k')
def Z_rot(j,T): #input starting j and go up in integer steps to j+10
    b=1./(T*8.617e-5) #boltzmann const in ev/K
    erot=0.0076 #eV this is for diatomic hydrogen
    a=(2.*j+1.)*np.exp(-b*(j*(j+1.))*erot)+(2.*(j+1.)+1.)*np.exp(-b*((j+1.)*((j+1.)+1.))*erot)+(2.*(j+2.)+1.)*np.exp(-b*((j+2.)*((j+2.)+1.))*erot)+(2.*(j+3.)+1.)*np.exp(-b*((j+3.)*((j+3.)+1.))*erot)+(2.*(j+4.)+1.)*np.exp(-b*((j+4.)*((j+4.)+1.))*erot)+(2.*(j+5.)+1.)*np.exp(-b*((j+5.)*((j+5.)+1.))*erot)+(2.*(j+6.)+1.)*np.exp(-b*((j+6.)*((j+6.)+1.))*erot)+(2.*(j+7)+1.)*np.exp(-b*((j+7.)*((j+7.)+1.))*erot)+(2.*(j+8.)+1.)*np.exp(-b*((j+8.)*((j+8.)+1.))*erot)+(2.*(j+9.)+1.)*np.exp(-b*((j+9.)*((j+9.)+1.))*erot)+(2.*(j+10.)+1.)*np.exp(-b*((j+10.)*((j+10.)+1.))*erot)
    return a
plt.plot(T_rot,Z_rot(0.,T_rot))
plt.xlim(xmax=5000)

plt.show()
```

```python
plt.figure(num=None, figsize=(8, 6), dpi=80, facecolor='w', edgecolor='k')

def ln_Z(T):
    return np.log(Z_rot(0,T))

def U_over_N(T):
    h=1
    a=(ln_Z(T+h/2.)-ln_Z(T-h/2.))/h
    return a*(T**2.)*8.617e-5

plt.xlim(0,5000)
plt.xlabel("$T$ (K)",fontsize=15)
plt.ylabel("$U/N$ (eV)", fontsize=15)
plt.plot(T_rot,U_over_N(T_rot))
plt.show()
```

```python
from numpy import log
def Ortho_Z_rot(j,T): #input starting j and go up in only odd integer steps to j+10 for orthohydrogen
    b=1./(T*8.617e-5) #boltzmann const in ev/K
    erot=0.0076 #eV this is for diatomic hydrogen
    a=(2.*(j+1.)+1.)*np.exp(-b*((j+1.)*((j+1.)+1.))*erot)+(2.*(j+3.)+1.)*np.exp(-b*((j+3.)*((j+3.)+1.))*erot)+(2.*(j+5.)+1.)*np.exp(-b*((j+5.)*((j+5.)+1.))*erot)+(2.*(j+7)+1.)*np.exp(-b*((j+7.)*((j+7.)+1.))*erot)+(2.*(j+9.)+1.)*np.exp(-b*((j+9.)*((j+9.)+1.))*erot)
    return a

def Para_Z_rot(j,T): #input starting j and go up in only even integer steps to j+10 for parahydrogren
    b=1./(T*8.617e-5) #boltzmann const in ev/K
    erot=0.0076 #eV this is for diatomic hydrogen
    a=(2.*j+1.)*np.exp(-b*(j*(j+1.))*erot)+(2.*(j+2.)+1.)*np.exp(-b*((j+2.)*((j+2.)+1.))*erot)+(2.*(j+4.)+1.)*np.exp(-b*((j+4.)*((j+4.)+1.))*erot)+(2.*(j+6.)+1.)*np.exp(-b*((j+6.)*((j+6.)+1.))*erot)+(2.*(j+8.)+1.)*np.exp(-b*((j+8.)*((j+8.)+1.))*erot)+(2.*(j+10.)+1.)*np.exp(-b*((j+10.)*((j+10.)+1.))*erot)
    return a

def Para_ln_Z(T):
    return np.log(Para_Z_rot(0,T))

def Para_U_over_N(T):
    h=.1
    a=(Para_ln_Z(T+h/2.)-Para_ln_Z(T-h/2.))/h
    return a*(T**2.)*8.617e-5

def Ortho_ln_Z(T):
    return log(Ortho_Z_rot(0,T))

def Ortho_U_over_N(T):
    h=.1
    a=(Ortho_ln_Z(T+h/2.)-Ortho_ln_Z(T-h/2.))/h
    return a*(T**2.)*8.617e-5

def Para_C_over_Nk(T):
    h=.1
    a=(Para_U_over_N(T+h/2.)-Para_U_over_N(T-h/2.))/h
    return a/8.617e-5
def Ortho_C_over_Nk(T):
    h=.1
    a=(Ortho_U_over_N(T+h/2.)-Ortho_U_over_N(T-h/2.))/h
    return a/8.617e-5
```

```python
def Tot_C_over_Nk(T):
    h=.1
    a=(Ortho_U_over_N(T+h/2.)-Ortho_U_over_N(T-h/2.))/h
    b=(Para_U_over_N(T+h/2.)-Para_U_over_N(T-h/2.))/h
    return .25*b/8.617e-5 + .75*a/8.617e-5
```
