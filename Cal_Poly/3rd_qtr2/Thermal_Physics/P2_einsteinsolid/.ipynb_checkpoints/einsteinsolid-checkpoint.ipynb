Consider a “solid” consisting of N identical oscillators (and thus N/3 atoms), each with energy 
quantized in chunks of size ε = hf, where f is the oscillator frequency. The total energy of the 
solid is U = qε, where q is the total number of chunks of energy of all oscillators. The oscillators 
share energy, but q cannot change unless the solid interacts with its environment.

Part 1
===

Consider a solid with N = 100. The entropy of the solid is S / k = lnΩ. That is, lnΩ is the 
entropy in units of k. Set up a spreadsheet in which the first column is q, the second is Ω, and the 
third is S/k. Start with 10 or so rows; you’ll expand the number of rows in a little while. 

```python
import numpy as np
import math
from math import factorial as fact
import csv
```

```python
def make_einsolid(N,q_lim):
    q = range(0,q_lim)

    Om=[]
    for i in q:
        Om.append(fact(1+N+i)/fact(i)/fact(N-1))
    
    Sk=[]
    for i in Om:
        Sk.append(math.log(i))
    
    kTe=[]
    for i in range(len(Sk)):
        if i==len(Sk)-1:
            break
        if i!=0:
            kTe.append((float(Sk[i+1]-Sk[i-1])/float(q[i+1]-q[i-1]))**(-1))
    
    q_cut=q[1:-1]
    CNk=[]
    for i in range(len(q_cut)):
        if i==len(q_cut)-1:
            break
        if i!=0:
            CNk.append((float(q_cut[i+1]-q_cut[i-1])/float(kTe[i+1]-kTe[i-1])))
    return (q,q_cut,Om,Sk,kTe,CNk)

N=100
q_lim=200
q,q_cut,Om,Sk,kTe,CNk=make_einsolid(N,q_lim)

Om_sci=[]
for i in Om: Om_sci.append('{:e}'.format(float(i)))

with open('part1_dat.csv','w') as csvfile:
    filewriter = csv.writer(csvfile)
    filewriter.writerow(['q','Omega','S/k','kT/e','q_cut','C/(Nk)'])
    filewriter.writerows(zip(q,Om_sci,Sk,kTe,q_cut,CNk))
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.plot(q,Om)
plt.show()
plt.plot(q,Sk)
plt.show()
plt.plot(q_cut,kTe)
plt.show()
```
