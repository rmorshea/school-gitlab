A “classic” problem in statistical physics is the random walk problem. It is directly related to 
many real world problem ranging from physics (diffusion of gases, electron motion in solids) to 
biology (various problems in genetics and ecology) to economics (price fluctuations). 
 
Although the basic random walk problem can be completely and exactly analyzed using 
probability theory, we’re going to study the problem “empirically” with the use of random 
numbers. Many physics problems that lack an exact solution can nonetheless be well understood 
by looking at the statistics of many randomly generated events. This statistical approach to 
understanding complex situations is called the Monte Carlo method, named for the Monte Carlo 
Casino in Monaco where fortunes are won and lost in games of chance. For those of you who 
return for Physics 401, we’ll use the Monte Carlo method to understand ferromagnetism.

```python
def rand_walk(L):
    counter=1
    step=[0]
    dist=[0]
    for i in range(L):
        step.append(counter)
        rand=randsign()
        if randsign==1.0:
            dist.append(dist[i]+rand)
        else:
            dist.append(dist[i]+rand)
        counter+=1
    return [step,dist]
```

```python
def step_plot(L,num_graphs,filename=None):
    plt.figure(figsize=(10,5))
    x=[]
    y=[]
    for i in range(num_graphs):
        x.append(rand_walk(L)[0])
        y.append(rand_walk(L)[1])
    for j in range(num_graphs):
        x=np.array(x)
        yabs=np.array(y)
        
        yabs=abs(yabs)
        maxend=yabs.max()
        yend=y[j][-1]
        
        plt.plot(x[j],y[j],color=str(abs(yend/1.25/(maxend))))
        plt.xlabel('Step')
        plt.ylabel('Distance')
        plt.grid(True)
        if filename!=None:
            plt.savefig(filename,bboxinches='tight')
    plt.show()
```

```python
step_plot(50,10,'50step.png')
step_plot(100,10,'100step.png')
step_plot(400,10,'400step.png')
step_plot(1000,10,'1000step.png')
```

```python
def avg_end(num,L):
    end=[]
    for i in range(num):
        end.append(rand_walk(L)[1][-1])
    return np.average(end)
```

```python
def ends_hist(num,L,plot=True,filename=None):
    '''num = number random walks which will be generated
    L = the length of the random walks
    filename = takes a string as the name of the file which the plot will be saved as'''
    
    def gaus(x,x0,sig,c):
        return ((c)/(np.sqrt(2*np.pi)*sig))*np.exp(-((x-x0)**2)/(2.*sig**2))
    from scipy.optimize import curve_fit
    
    end=[]
    for i in range(num):
        end.append(rand_walk(L)[1][-1])
    ran=max(end)-min(end)
    if plot==True:
        plt.figure(figsize=(10,5))
    counts,bin_edges = np.histogram(end,ran)
    bin_cents = (bin_edges[:-1] + bin_edges[1:])/2.
    counts=counts.tolist()
    bin_cents=bin_cents.tolist()
    j=0
    for i in range(len(counts)):
        if i-j==len(counts):
            break
        if counts[i-j]==0:
            counts.pop(i-j)
            bin_cents.pop(i-j)
            j+=1
    plsq,pcov=curve_fit(gaus,bin_cents,counts)
    if plot==True:
        x = np.linspace(min(end)*1.01,max(end)*1.01,1000)
        y = []
        for i in x:
            y.append(gaus(i,plsq[0],plsq[1],plsq[2]))
        plt.plot(x,y,color='r',label='gaussian fit')
        plt.scatter(bin_cents,counts,marker='.',label='random walk distances (binned)')
        plt.xlim(min(end)*1.01,max(end)*1.01)
        plt.ylim(0,max(counts)*1.1)
        plt.xlabel('Distance')
        plt.ylabel('Number')
        plt.legend(loc='best')
        if filename!=None:
            plt.savefig(filename, bbox_inches='tight')
        plt.show()
    return plsq
```

```python
nums_i=[50,100,400,900]
nums=[]
sigs=[]
for i in nums_i:
    for j in range(4):
        sigs.append(ends_hist(1000,i,False)[1])
        nums.append(i)
        
print sigs

def sqrt(x,a,b):
    return a*np.sqrt(x-b)
from scipy.optimize import curve_fit

plsq,pcov=curve_fit(sqrt,nums,sigs)

x=np.linspace(0,950,1000)
y=[]
for i in x:
    y.append(sqrt(i,plsq[0],plsq[1]))

plt.figure(figsize=(10,5))
plt.plot(x,y,color='r',label='sqrt fit function')
plt.scatter(nums,sigs,marker='.',label='sigma vs steps')
plt.xlim(0,950)
plt.ylim(0,35)
plt.xlabel('steps')
plt.ylabel('sigma')
plt.legend(loc='upper left')
plt.savefig('sqrt_fit.png',bbox_inches='tight')
plt.show()
```

```python
def factorial(N):
    if N<0 or type(N)!=int:
        raise ValueError('N must be a positive integer')
    fact=1
    for i in range(N+1)[1:]:
        fact=i*fact
    return fact
```
