```python
ptime = dat851 = np.genfromtxt('opt-pump_dat/scope_1.txt',delimiter=',',skiprows=10,usecols=[0,2])
```

```python
ptime_t=[]
ptime_v=[]
for i in ptime:
    ptime_t.append(i[0])
    ptime_v.append(i[1])
plt.plot(np.array(-np.array(ptime_t[550:950])),-np.array(ptime_v[550:950]))
ptime_t=np.array(ptime_t[550:900])
ptime_v=np.array(ptime_v[550:900])
plt.show()
```

```python
def exp(t,T,b,B):
    return np.e**((t-b)*T)+B
plsq,pcov = curve_fit(exp,-ptime_t,-ptime_v)
print plsq
print np.sqrt(np.abs(pcov))
```

```python
xran=np.linspace(0,0.045,100)
yran=[]
for i in xran:
    yran.append(exp(i,plsq[0],plsq[1],plsq[2]))
plt.figure(figsize=(10,5))
plt.plot(np.array(xran),np.array(yran),color='r')
plt.scatter(-ptime_t,-ptime_v,color='k',marker='.',linewidths=0.2)
plt.xlim(0.008,0.045)
plt.xlabel('time (s)')
plt.ylabel('voltage (V)')
plt.savefig('exp_fit.png',bboxinch='tight')
plt.show()
```
