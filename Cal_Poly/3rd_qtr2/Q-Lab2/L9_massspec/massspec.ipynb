```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
Ar_dat = zip(*np.loadtxt('data/d2argon.txt',delimiter=',',usecols=(0,1),skiprows=22))
Atm_dat = zip(*np.loadtxt('data/d2atmosphere.txt',delimiter=',',usecols=(0,1),skiprows=22))
Bre_dat = zip(*np.loadtxt('data/d2breath.txt',delimiter=',',usecols=(0,1),skiprows=22))
Eth_dat = zip(*np.loadtxt('data/d2ethanol.txt',delimiter=',',usecols=(0,1),skiprows=22))
He_dat = zip(*np.loadtxt('data/d2helium.txt',delimiter=',',usecols=(0,1),skiprows=22))
Kr_dat = zip(*np.loadtxt('data/d2krypton.txt',delimiter=',',usecols=(0,1),skiprows=22))
Ne_dat = zip(*np.loadtxt('data/d2neon.txt',delimiter=',',usecols=(0,1),skiprows=22))
Ni_dat = zip(*np.loadtxt('data/d2nitrogen.txt',delimiter=',',usecols=(0,1),skiprows=22))
Nos_dat = zip(*np.loadtxt('data/d2nosample.txt',delimiter=',',usecols=(0,1),skiprows=22))
Data_list=[Atm_dat,Bre_dat,Ar_dat,Eth_dat,He_dat,Ne_dat,Ni_dat,Nos_dat]
Title_list=['Atmosphere','Breath','Argon','Ethanol','Hellium','Neon','Nitrogen','Residual']
plt.bar(Kr_dat[0],Kr_dat[1],label='Krypton')
plt.ylim(ymin=0)
plt.xlim(0,85)
plt.legend()
plt.show()
plt.bar(Nos_dat[0],Nos_dat[1],label='Residual')
plt.ylim(ymin=0)
plt.xlim(0,85)
plt.legend()
plt.show()
```

```python
Norm_pdat=[]
Norm_mdat=[]
for i in range(len(Data_list)):
    Norm_pdat.append(np.array(Data_list[i][1]))#-np.array(Nos_dat[1]))
    Norm_mdat.append(Data_list[i][0])
for i in range(len(Data_list)-1):
    plt.bar(range(1,len(Norm_mdat[i])+1),Norm_pdat[i],label=Title_list[i])
    plt.ylim(ymin=0)
    plt.xlim(0,65)
    plt.legend()
    plt.show()
    
plt.bar(Kr_dat[0],Kr_dat[1],label='Krypton')
plt.ylim(ymin=0)
plt.xlim(0,65)
plt.legend()
plt.show()
plt.bar(Nos_dat[0],Nos_dat[1],label='Residual')
plt.ylim(ymin=0)
plt.xlim(0,65)
plt.legend()
plt.show()
```
