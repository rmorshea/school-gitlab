```python
import numpy as np
disc=np.array([5,10,15,20,30,40,50,60,80,100,120]) #discrimator level. unit: mV. unc: +/-0.1 mV
disc=-disc
cnts=[8588,7827,7046,6492,4360,2583,1399,360,24,11,7]
dkcnts=[421,143,174,164,119,58,28,11,2,0,0]

print len(cnts)
print len(disc)
print len(dkcnts)

CntBkg_rat=[]
for i in range(len(cnts)):
    CntBkg_rat.append(float(dkcnts[i])/cnts[i])
```

```python
import matplotlib.pyplot as plt

plt.Figure(figsize=(10,10))
plt.scatter(disc,cnts,marker='.')
plt.show()

plt.scatter(disc,dkcnts)
plt.show()

plt.scatter(disc,CntBkg_rat,marker='.')
plt.plot(disc,CntBkg_rat,linewidth=0.5,color='r')
plt.show()
```

```python
dat3000c = np.loadtxt('phot_dat1.csv',delimiter=',',usecols=[0])
dat1000c = np.loadtxt('phot_dat1.csv',delimiter=',',usecols=[2])
dat10000c = np.loadtxt('phot_dat1.csv',delimiter=',',usecols=[4])

dat3000 = np.loadtxt('phot_dat.csv',delimiter=',',usecols=range(10))
dat1000 = np.loadtxt('phot_dat.csv',delimiter=',',usecols=range(12,22))
dat10000 = np.loadtxt('phot_dat.csv',delimiter=',',usecols=range(23,33))
```

```python
dat3kc=[]
dat1kc=[]
dat10kc=[]
for i in dat3000c:
    dat3kc.append(i)
for i in dat1000c:
    dat1kc.append(i)
for i in dat10000c:
    dat10kc.append(i)
```

```python
dat3k=[]
dat1k=[]
dat10k=[]
for i in dat3000:
    for j in i:
        dat3k.append(j)
for i in dat1000:
    for j in i:
        dat1k.append(j)
for i in dat10000:
    for j in i:
        dat10k.append(j)
```

```python
import matplotlib as mpltb
import math
counts,bin_edges = np.histogram(dat3kc,int(max(dat3kc)))
bin_cents = bin_edges[:-1]+0.5

x=np.array(range(0,int(max(dat3kc))))
y=[]
avg=np.average(dat3kc)
for i in x:
    y.append((avg**i)/(math.factorial(i)*(np.e)**(avg)))
x_cents=x

er=np.sqrt(counts)/1100.

counts=counts/1100.

plt.figure(figsize=(25,12.5))
plt.errorbar(bin_cents+0.25,counts,er,fmt=' ',color='k',linewidth=2)
plt.bar(x_cents,y,0.4,color='b',label='theory')
plt.bar(bin_cents,counts,0.4,color='y',label='raw data')
plt.xlim(0,10)
plt.ylim(0,0.32)
plt.legend(prop={'size':30})
mpltb.rc('xtick', labelsize=30)
mpltb.rc('ytick', labelsize=30)
plt.ylabel('$P(n,T)$',fontsize=35)
plt.xlabel('$n$ (counts)',fontsize=35)

plt.savefig('3kc.png',bboxinches='tight')
plt.show()

chi=0
for i in range(len(y)):
    if er[i]!=0 and np.math.isnan(y[i])!=True:
        chi+=(counts[i]-y[i])**2/er[i]**2

print 'chi squared =',chi/int(max(dat3kc))
```
