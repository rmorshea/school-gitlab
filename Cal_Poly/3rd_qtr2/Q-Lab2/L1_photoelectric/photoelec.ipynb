```python
c = 3e8
e = 1.602e-19
wav = [c/365.0e-9,c/404.7e-9,c/435.8e-9,c/546.1e-9,c/578.0e-9]
Vs = [Vs365_avg,Vs404_avg,Vs435_avg,Vs546_avg,Vs578_avg]

def lin_fit(fre,A,B):
    return -A*fre+B

from scipy.optimize import curve_fit
p0 = [5.76778884263e-25/e,1.8977742975e-19/e]
plsq,pcov = curve_fit(lin_fit,array(wav),array(Vs),p0,0.001*ones(len(wav)))

print plsq

x = linspace(c/350e-9,c/600e-9,1000)
y = []
y2 = []
for i in x:
    y.append(lin_fit(i,plsq[0],plsq[1]))
y = array(y)

y3 = []
for i in x:
    y3.append(lin_fit(i,4.1360799001248435e-15,plsq[1]))


print sqrt(pcov)

figsize(10,7)
plot(x,y,color='b',label='${A}{f}+B$')
plot(x,y3,color='r', label='$hf+B$')
scatter(wav,Vs,color='g',label='Raw Data')
xlim(c/360e-9,c/585e-9)
ylim(-1.8,-0.6)
xlabel('Frequency (Hz)',fontsize=13)
ylabel('Stopping Voltage (V)',fontsize=13)
legend(loc='upper left')
show()
```

```python
h = plsq[0]*e
w = plsq[1]*e
dh = sqrt(abs(pcov[0][0]))*e
dw = sqrt(abs(pcov[1][1]))*e
f_t = w/h
df_t = sqrt((dw/h)**2+(w/(h**2)*dh)**2)

print 'h =',h,'+/-',dh,'J*s'
print 'W =',w,'+/-',dw,'J'
print 'threshold frequency =',f_t,'+/-',df_t,'Hz'
```
