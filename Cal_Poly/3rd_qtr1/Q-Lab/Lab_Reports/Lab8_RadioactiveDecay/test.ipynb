```python
tau_dead=400e-6
```

```python
Ag_dat=loadtxt('SilverData.csv',unpack=True, delimiter=',',skiprows=6,usecols=[0,1])
#print silver_dat[1]
figsize(9,6)
plot(range(256),Ag_dat[1])
#print shape(silver_dat[1])
xlim(0,250)
yscale('log')
ylim(0,1000)
show()
```

```python
Rcounts_Ag = Ag_dat[1] #raw counts

Tcounts_Ag = [] #true counts
for i in range(shape(Rcounts_Ag)[0]):
    Tcounts_Ag.append(Rcounts_Ag[i]/(1.-(Rcounts_Ag[i]*(tau_dead/4.))))
```

```python
dRcounts_Ag = []
for i in Rcounts_Ag:
    dRcounts_Ag.append(sqrt(i))

dTcounts_Ag = [] #error for log(Tcounts_Ag)
for i in range(len(Rcounts_Ag)):
    A = (dRcounts_Ag[i]/(1-float(Rcounts_Ag[i]*tau_dead)/4.))/(Rcounts_Ag/(1-Rcounts_Ag*100e-6))
    B = ((Rcounts_Ag[i]*float(dRcounts_Ag[i]*tau_dead)/4.)/(1-float(Rcounts_Ag[i]*tau_dead)/4.)**2)/(Rcounts_Ag/(1-Rcounts_Ag*100e-6))
    dTcounts_Ag.append(sqrt(A**2+B**2))
```
