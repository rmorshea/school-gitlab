```python
def dRcount_list(Rcounts_Ag):
    dRcounts_Ag = []
    for i in Rcounts_Ag:
        dRcounts_Ag.append(sqrt(i))
    return dRcounts_Ag

def logRcount_list(Tcounts_Ag):
    log_Ag=[]
    for i in Tcounts_Ag:
        log_Ag.append(log(i))
    return log_Ag

def dTcount_list(Rcounts_Ag,dRcounts_Ag,tau_dead):
    dTcounts_Ag = [] #error for log(Tcounts_Ag)
    for i in range(len(Rcounts_Ag)):
        A = (dRcounts_Ag[i]/(1-float(Rcounts_Ag[i]*tau_dead)))
        B = ((Rcounts_Ag[i]*float(dRcounts_Ag[i]*tau_dead))/(1-float(Rcounts_Ag[i]*tau_dead))**2)
        dTcounts_Ag.append(sqrt(A**2+B**2))
    return dTcounts_Ag
        
def dlogTcount_list(Rcounts_Ag,dRcounts_Ag,tau_dead):
    dlog_Ag = [] #error for log(Tcounts_Ag)
    for i in range(len(Rcounts_Ag)):
        A = (dRcounts_Ag[i]/(1-float(Rcounts_Ag[i]*tau_dead)))/(Rcounts_Ag[i]/(1-Rcounts_Ag[i]*tau_dead))
        B = ((Rcounts_Ag[i]*float(dRcounts_Ag[i]*tau_dead))/(1-float(Rcounts_Ag[i]*tau_dead))**2)/(Rcounts_Ag[i]/(1-Rcounts_Ag[i]*tau_dead))
        dlog_Ag.append(sqrt(A**2+B**2))
    return dlog_Ag
```

```python
dRcounts_Ag1 = dRcount_list(Rcounts_Ag1N)
dTcounts_Ag1 = dTcount_list(Rcounts_Ag1N,dRcounts_Ag1,tau_dead)
log_Ag1 = logRcount_list(Tcounts_Ag1)
dlog_Ag1 = dlogTcount_list(Rcounts_Ag1N,dRcounts_Ag1,tau_dead)
```

```python
def log_part(t,B,tau2,C):
    'returns the sum of two exponentials and a constant'
    return ln(B*exp(-(t)/tau2)+C)

def exp_part(t,B,tau2,C):
    'returns the sum of two exponentials and a constant'
    return B*exp(-(t)/tau2)+C

def exp_comp(t,A,tau1,B,tau2,C):
    'returns the sum and individual components of exp_sum in a list'
    return [A*exp(-t/tau1)+B*exp(-t/tau2)+C, A*exp(-t/tau1), B*exp(-t/tau2), C]
```

```python
from scipy.optimize import curve_fit
x_Ag = array(linspace(80,105,25))
y_Ag = array(Tcounts_Ag1[80:105])
d_Ag = array(dTcounts_Ag1[80:105])
p0 = array([120,200,1])
```
